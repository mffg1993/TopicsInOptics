# Notebook: Vector Beams – Definition, Cylindrical Vector Beams, and Full Poincaré Beams

---

## 1. Definition

**Vector beams** are light fields whose **polarization varies across the transverse plane**.  
Unlike scalar beams (where the polarization direction is uniform), vector beams feature a spatially structured polarization that can be tightly coupled with the beam’s amplitude and phase.

### 1.1 Mathematical Form

A general vector beam can be expressed as:
\[
\mathbf{E}(x,y) = E_x(x,y) \, \hat{\mathbf{x}} + E_y(x,y) \, \hat{\mathbf{y}}
\]

- \(E_x(x,y)\) and \(E_y(x,y)\): complex amplitudes of the field along the \(x\)- and \(y\)-polarization directions.
- Polarization is **position-dependent**: direction and ellipticity change across the transverse plane.

### 1.2 Properties

- Possess **spatially varying polarization**.
- Can combine **scalar spatial modes** (e.g., Gaussian, LG) with polarization structure.
- Enable unique light-matter interactions, tighter focusing, and polarization singularities.

## 2. Cylindrical Vector Beams

### 2.1 Definition

**Cylindrical vector beams (CVBs)** are a special class of vector beams where the polarization is defined in **cylindrical coordinates**.  
The two canonical examples are:

1. **Radially polarized beam**:
   \[
   \mathbf{E}_{\text{radial}}(r,\phi) = E_0(r) \, \hat{r}
   = E_0(r) \left( \cos\phi \, \hat{\mathbf{x}} + \sin\phi \, \hat{\mathbf{y}} \right)
   \]

2. **Azimuthally polarized beam**:
   \[
   \mathbf{E}_{\text{azimuthal}}(r,\phi) = E_0(r) \, \hat{\phi}
   = E_0(r) \left( -\sin\phi \, \hat{\mathbf{x}} + \cos\phi \, \hat{\mathbf{y}} \right)
   \]

### 2.2 Properties

- Symmetry is defined around the beam axis.
- Radially polarized beams can be focused to smaller spots (useful in microscopy).
- Azimuthally polarized beams generate strong longitudinal magnetic fields when tightly focused.

### 2.3 Higher-Order CVBs

- Introduce orbital angular momentum (OAM) by multiplying the transverse field with \(e^{i l \phi}\).
- Results in **vector vortex beams** with combined polarization and phase singularities.

## 3. Full Poincaré Beams

### 3.1 Definition

**Full Poincaré (FP) beams** are vector beams that **cover all possible polarization states of the Poincaré sphere** within a single transverse plane.

- Include **linear, circular, and elliptical polarizations** distributed across the beam.
- Typically generated as a **superposition** of orthogonal circular polarization states with different spatial modes.

### 3.2 Mathematical Form

One way to write an FP beam:
\[
\mathbf{E}_{\text{FP}}(x,y) =
c_1 \, u_1(x,y) \, \hat{\mathbf{e}}_L +
c_2 \, u_2(x,y) \, \hat{\mathbf{e}}_R
\]

- \(u_1, u_2(x,y)\): spatial modes (e.g., Gaussian + Laguerre–Gaussian).
- \(\hat{\mathbf{e}}_L, \hat{\mathbf{e}}_R\): left and right circular polarizations.
- Coefficients \(c_1, c_2\) define the balance of each component.

### 3.3 Properties

- Include polarization singularities (points of undefined polarization).
- Rich topology: **C-points** (circular polarization points) and **L-lines** (linear polarization lines).
- Useful for **polarization-based metrology** and advanced light-matter interactions.

## 4. Visualization

### 4.1 Polarization Maps

We can represent vector beams using:
1. **Intensity map:** \(I(x,y) = |E_x|^2 + |E_y|^2\)
2. **Quiver plot (arrows):** arrows show the local polarization direction.

### 4.2 Examples

- **Radial beam:** arrows point outward from the center.
- **Azimuthal beam:** arrows form closed circles.
- **Full Poincaré beam:** arrows vary smoothly, covering all polarization states.

---

## 5. Summary

- Vector beams exhibit spatially varying polarization, unlike scalar beams.
- Cylindrical Vector Beams (radial, azimuthal) are the most common examples.
- Full Poincaré beams contain **all possible polarization states** across the beam.
- Visualization is best done with intensity maps combined with polarization direction arrows.
