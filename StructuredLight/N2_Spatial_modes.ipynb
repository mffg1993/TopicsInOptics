# Notebook: Higher-Order and Exotic Spatial Modes of Light

---

## 1. Introduction

Higher-order and exotic spatial modes extend beyond the fundamental Gaussian beam and the basic Hermite–Gaussian (HG) and Laguerre–Gaussian (LG) families.  
These modes exhibit complex amplitude and phase structures, enabling powerful applications in:
- Quantum optics and optical communications (orthogonal mode sets and orbital angular momentum)
- Microscopy, imaging, and structured illumination
- Optical trapping, particle manipulation, and nonlinear optics

This notebook covers:
1. **Hermite–Gaussian (HG)** modes – rectangular symmetry
2. **Laguerre–Gaussian (LG)** modes – cylindrical symmetry, orbital angular momentum
3. **Elegant Laguerre–Gaussian (eLG)** modes – alternative LG definition
4. **Ince–Gaussian (IG)** modes – elliptical symmetry
5. **Mathieu beams** – exact Helmholtz solutions in elliptical coordinates
6. **Regular Bessel beams** – ideal non-diffracting beams
7. **Bessel–Gaussian (BG) beams** – finite-energy approximations of Bessel beams
8. **Airy beams** – accelerating and self-healing beams

---

## 2. Hermite–Gaussian (HG) Modes

### 2.1 Description
Hermite–Gaussian modes have **rectangular symmetry** and are indexed by \((m,n)\):  

\[
HG_{mn}(x,y,z) = HG_m(x,z) HG_n(y,z)
\]

where:
\[
HG_m(x,z) = \sqrt{\frac{2}{\pi}} \frac{1}{w(z)}
H_m\left( \frac{\sqrt{2}x}{w(z)} \right)
\exp\left( -\frac{x^2}{w^2(z)} \right)
\]

- \(H_m(x)\): Hermite polynomial  
- Orthogonal basis functions in rectangular coordinates.

### 2.2 Visualization
- Intensity profiles show nodal lines along \(x\) and \(y\).
- Phase alternates between adjacent lobes.

---

## 3. Laguerre–Gaussian (LG) Modes

### 3.1 Standard LG Modes
These modes have **cylindrical symmetry** and are indexed by radial \((p)\) and azimuthal \((l)\) indices:

\[
LG_p^l(r,\phi,z) = C \, r^{|l|}
L_p^{|l|} \left( \frac{2r^2}{w^2(z)} \right)
\exp\left( -\frac{r^2}{w^2(z)} \right)
\exp(i l \phi)
\]

- \(L_p^{|l|}(x)\): associated Laguerre polynomial  
- \(l \neq 0\): azimuthal phase singularity and orbital angular momentum (OAM)  

### 3.2 Elegant Laguerre–Gaussian (eLG) Modes
The **Elegant LG** formulation uses complex beam parameters and is often used for Fourier optics and propagation studies:

\[
eLG_p^l(r,\phi,z) = \frac{1}{q(z)}
\left( \frac{r \sqrt{2}}{q(z)} \right)^{|l|}
L_p^{|l|}\left( \frac{2r^2}{q^2(z)} \right)
\exp \left( - \frac{r^2}{q(z)} \right) e^{i l \phi}
\]

- \(q(z) = z - i z_R\) is the complex beam parameter.
- Reduces to standard LG when written in real beam parameters \(w(z), R(z)\).

## 4. Ince–Gaussian (IG) Modes

### 4.1 Description
IG modes are solutions of the paraxial wave equation in **elliptical coordinates**, indexed by order \(p\), degree \(m\), and parity (even or odd):

\[
IG_{p,m}^{e/o}(x,y,z) = C \,
\text{Ce}_m^p(\xi,\epsilon) \,
\text{Ce}_m^p(\eta,\epsilon)
\]

- \(\text{Ce}_m^p\) and \(\text{Se}_m^p\): even/odd Ince polynomials  
- \(\epsilon\): ellipticity parameter  

### 4.2 Properties
- Connects HG and LG families:  
  - \(\epsilon \to 0\) → LG  
  - \(\epsilon \to \infty\) → HG  

---

## 5. Mathieu Beams

### 5.1 Description
Mathieu beams are **exact solutions** of the Helmholtz equation in elliptical coordinates:

\[
E(\xi, \eta) =
\text{Ce}_m(\xi,q) \,
\text{Ce}_m(\eta,q)
\]

- \(\text{Ce}_m, \text{Se}_m\): Mathieu functions  
- \(q\): ellipticity parameter  

### 5.2 Properties
- Non-diffracting and self-healing beams.
- Intricate elliptical intensity distributions.

---

## 6. Regular Bessel Beams

### 6.1 Description
Ideal non-diffracting beams that are exact solutions to the Helmholtz equation:

\[
E(r,z) = J_0(k_r r) e^{i k_z z}
\]

- \(J_0(x)\): zeroth-order Bessel function  
- \(k_r^2 + k_z^2 = k^2\)

### 6.2 Properties
- Central bright core with concentric rings.
- Perfectly non-diffracting (require infinite energy).

---

## 7. Bessel–Gaussian (BG) Beams

### 7.1 Description
Finite-energy approximation to the ideal Bessel beam:

\[
E(r,z) = J_0(k_r r)
\exp\left( -\frac{r^2}{w_0^2} \right)
\exp(i k_z z)
\]

- Gaussian envelope truncates the rings.

### 7.2 Properties
- Retain partial diffraction-free behavior over a finite propagation distance.

---

## 8. Airy Beams

### 8.1 Description
Accelerating and self-healing beams described by the Airy function:

\[
E(x,z) = Ai \left( \frac{x}{x_0} - \frac{z^2}{4k^2 x_0^4} \right)
\exp \left( i \left( \frac{x z}{2 k x_0^3} - \frac{z^3}{12 k^3 x_0^6} \right) \right)
\]

- \(Ai(x)\): Airy function  
- \(x_0\): beam scale  

### 8.2 Properties
- The main lobe follows a parabolic trajectory.
- Beam can reconstruct after partial obstruction (**self-healing**).
